Missing Values

```python
##Missing values in categorical features
##checking for column having null values

cat_missing_feature = [feature for feature in df.columns if df[feature].dtypes == 'O' and df[feature].isnull().sum() > 1]

cat_missing_feature
```

```python
def replace_cat_nan(df,feature_nan):
    data = df.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

df = replace_cat_nan(df,cat_missing_feature)
```

```python
df[cat_missing_feature].isnull().sum()
```

```python
df
```

```python
##Missing values in numerical feature

num_missing_feature = [feature for feature in df.columns if df[feature].dtypes != 'O' and df[feature].isnull().sum() >1]

df[num_missing_feature].isnull().sum()
```

```python
##Replacing nan values in numerical column with median as there are many outliers

for feature in num_missing_feature:
    med = df[feature].median()
    
    ##creating extra column for more info about nan value
    df[feature+'_nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(med,inplace=True)
    
    
```

```python
df[num_missing_feature].isnull().sum()
```

```python
##Temperal feature

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature] = df['YrSold']-df[feature]
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']]
```

Numerical Feature

Since the numerical feature of our dataset is skewed, there is need to perform log normal distribution

```python
##taking some skewed nuemrical column
num_feature = ['LotFrontage','LotArea','1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    df[feature] = np.log(df[feature])
```

```python
df[num_feature]
```

Handling rare values in categorical feature

##if the feature has less than 1% record, then it will be a rare category

```python
cat_feature = [feature for feature in df.columns if df[feature].dtypes == 'O']

cat_feature
```

```python
for feature in cat_feature:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp1 = temp[temp>0.1].index
    df[feature] = np.where(df[feature].isin(temp1),df[feature],'Rare_var')
    
```

```python
df
```

Feature Scaling

```python
for feature in cat_feature:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df
```

```python
scale_feature = [feature for feature in df.columns if feature not in ['Id','SalePrice'] ]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[scale_feature])
```

```python
##transform the train and test set, add on the id,saleprice and category feature as variable

data = pd.concat([df[['Id','SalePrice']].reset_index(drop=True),
                   pd.DataFrame(scaler.transform(df[scale_feature]), columns=scale_feature)],
                   axis=1)
```
